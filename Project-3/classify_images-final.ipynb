```python
import numpy as np
import keras
import tensorflow as tf
import pandas as pd
import os
from keras.applications.vgg16 import VGG16,preprocess_input, decode_predictions
from keras.preprocessing.image import load_img,img_to_array
import numpy as np
import glob
from shutil import copyfile
os.chdir(BASE_DIR)
!tree
```

```python
all_images = glob.glob("./Data/Base_Images/*.jpg") + glob.glob("./Data/Base_Images/*.png")
len(all_images)
```

```python
model = VGG16(include_top = True)
```

```python
model.summary()
```

```python
def get_the_top_prediction(path_to_image):
    im = load_img(path_to_image, target_size=(224, 224, 3))
    im=img_to_array(im)
    im=np.expand_dims(im,axis=0)
    im=preprocess_input(im)
    preds = model.predict(im)
    decoded_prediction = decode_predictions(preds, top = 1)[0][0][1]
    return decoded_prediction
```

```python
image_details = []
for image_path in all_images:
    print(len(image_details))
    prediction = get_the_top_prediction(image_path)
    image_details.append({
      'path' : image_path,
      'class' : prediction
    })
```

```python
set([details['class'] for details in image_details])
```

```python
# class from model : folder name in Project statement
folder_names = {
    'Arabian_camel' : 'arabian',
    'Chihuahua' : 'chihuahua',
    'German_shepherd': 'german',
    'acoustic_guitar': 'acoustic',
    'airliner': 'airliner',
    'church': 'church',
    'laptop': 'laptop',
    'tabby': 'tabby',
    'violin':'violin'
}
```

```python
df = pd.DataFrame(image_details)
```

```python
df.head()
```

```python
BASE_DIRECTORY_FOR_CLASSIFIED_IMAGES = "./Data/Classified_images/"
```

```python
df['folder'] = df['class'].apply(lambda x : folder_names[x])
df['file'] = df['path'].apply(lambda x : os.path.basename(x))

df['final_path'] = df.apply(lambda x :  BASE_DIRECTORY_FOR_CLASSIFIED_IMAGES + x.folder + "/" + x.file, axis = 1)
```

```python
df[['file', 'folder']].to_csv('./Data/image_folders.csv',index = False)
df.to_csv('./Data/complete.csv')
```
